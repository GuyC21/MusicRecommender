# The Actual Algorithem

```python
#saving the album_id
album_id = MergedData.loc[((MergedData['artistName'] == artistName) & (MergedData["albumName"]== albumName))]['album_id'].values[0]
```

```python
album_id
```

```python
#saving the data of our album so later we can do conditional indexing based on it
condition = MergedData.loc[(MergedData['artistName'].str.contains(artistName)) & (MergedData['album_id']==album_id)]
```

```python
"""
Sections:
1. Creating a Data grouped by albums.
2. Using the condition we created previously in order to keep only the data that answers this conditions:
    a. only values between 0.8x->1.5x (x being our album value for each column)
    b. keeping only values from 0.8x to 1.2x
    c. only values between 0.9x->1.5x (x being our album value for each column)
"""

#1
dfMean = MergedData.groupby(by=['album_id','albumName','artistName']).mean(numeric_only=True)
dfMean = dfMean.drop('duration',1)
#2a
dfMean = dfMean[(dfMean['instrumentalness']>=((condition.mean(numeric_only=True)['instrumentalness'])*0.80)) & (dfMean['instrumentalness']<=((condition.mean(numeric_only=True)['instrumentalness'])*1.5))]
dfMean = dfMean[(dfMean['danceability']>=((condition.mean(numeric_only=True)['danceability'])*0.80)) & (dfMean['danceability']<=((condition.mean(numeric_only=True)['danceability'])*1.5))]
dfMean = dfMean[(dfMean['energy']>=((condition.mean(numeric_only=True)['energy'])*0.80)) & (dfMean['energy']<=((condition.mean(numeric_only=True)['energy'])*1.5))]
dfMean = dfMean[(dfMean['speechiness']>=((condition.mean(numeric_only=True)['speechiness'])*0.80)) & (dfMean['speechiness']<=((condition.mean(numeric_only=True)['speechiness'])*1.5))]
dfMean = dfMean[(dfMean['acousticness']>=((condition.mean(numeric_only=True)['acousticness'])*0.80)) & (dfMean['acousticness']<=((condition.mean(numeric_only=True)['acousticness'])*1.5))]
dfMean = dfMean[(dfMean['liveness']>=((condition.mean(numeric_only=True)['liveness'])*0.80)) & (dfMean['liveness']<=((condition.mean(numeric_only=True)['liveness'])*1.5))]
dfMean = dfMean[(dfMean['valence']>=((condition.mean(numeric_only=True)['valence'])*0.80)) & (dfMean['valence']<=((condition.mean(numeric_only=True)['valence'])*1.5))]
#2b
dfMean = dfMean[(dfMean['mode']>=((condition.mean(numeric_only=True)['mode'])*0.80)) & (dfMean['mode']<=((condition.mean(numeric_only=True)['mode'])*1.2))]
#2c
dfMean = dfMean[(dfMean['tempo']>=((condition.mean(numeric_only=True)['tempo'])*0.9)) & (dfMean['tempo']<=((condition.mean(numeric_only=True)['tempo'])*1.5))]
dfMean['mean'] = dfMean[['mode','danceability','energy','speechiness','acousticness','instrumentalness','liveness','valence','tempo']].mean(axis=1)
```

```python
"""
After several QAs, ive came to the conclusion that popularity column is important in order to make the algorithem more consistent
In each condition, we are check the amount of results we got by the condition, and if its not enough, we 'lower the standards'
"""

#at least 50% as popular
conditional1 = dfMean[(dfMean['popularity']>=((MergedData.loc[(MergedData['artistName'].str.contains(artistName)) & (MergedData['albumName']==albumName)].mean(numeric_only=True)['popularity'])*0.5))]

#if none of the conditions met, do nothing
if len(conditional1.index)>=8:
    dfMean = conditional1
    print('conditional1 met')
else:
    #at least 30% as popular
    conditional2 = dfMean[(dfMean['popularity']>=((MergedData.loc[(MergedData['artistName'].str.contains(artistName)) & (MergedData['albumName']==albumName)].mean(numeric_only=True)['popularity'])*0.3))]
    if len(conditional2.index)>=8:
        dfMean = conditional2
        print('conditional2 met')
    else: 
        #at least 20% as popular
        conditional3 = dfMean[(dfMean['popularity']>=((MergedData.loc[(MergedData['artistName'].str.contains(artistName)) & (MergedData['albumName']==albumName)].mean(numeric_only=True)['popularity'])*0.2))]
        if len(conditional3.index)>=8:
            dfMean = conditional3
            print('conditional3 met')
        else:
            print("no conditional met")

#needs better explaination
if MergedData.loc[((MergedData['artistName']== artistName) & (MergedData["albumName"]== albumName))]['popularity'].iloc[0] != 0:
    # (mean*popularity) / mean+popularity
    dfMean['grade'] = dfMean['popularity']*.3+dfMean['mean']*.7
    #original code starts here
    dfMean = dfMean.sort_values(by='grade',ascending=False)
else:
    dfMean = dfMean.sort_values(by='mean',ascending=False)
```
